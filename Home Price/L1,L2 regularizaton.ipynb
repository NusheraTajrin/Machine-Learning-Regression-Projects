```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','CouncilArea','Propertycount','Regionname','Price']
```

```python
df=df[cols_to_use]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
fill_zero=['Distance','Bedroom2','Bathroom','Car','Propertycount']
df[fill_zero]=df[fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df=pd.get_dummies(df,drop_first=True)
df=df.astype(int)
df.head()
```

```python
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

Lasso Regression - L1

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=1000,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

Ridge Regression - L2

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
